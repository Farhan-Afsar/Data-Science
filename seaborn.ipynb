```python
"""
Statistical plot in seaborn
1.Relational Plot(relplot) -> Relation between two variable/feature
2.Distribution Plot(distplot)
3.Categorical Plot(catplot)
"""
```

```python
import seaborn as sns
sns.set_style('darkgrid')
sns.__version__
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
""" Ask and verify
New to ask question and verify using visualization
"""
tips.info()
```

```python
tips['time'].value_counts()
```

```python
tips['day'].value_counts()
```

```python
tips['smoker'].value_counts()
```

```python
tips['sex'].value_counts()
```

```python
sns.relplot(
    data=tips,
    x= 'total_bill',
    y = 'tip',
)
```

```python
sns.relplot(
    data=tips,
    x= 'total_bill',
    y = 'tip',
    hue = "day"
)
```

```python
sns.relplot(
    data=tips,
    x= 'total_bill',
    y = 'tip',
    hue = "day",
    row = "day"
)
```

```python
sns.relplot(
    data=tips,
    x = 'total_bill',
    y = 'tip',
    hue = "sex",
    col = "time"
)
```

```python
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue="sex",
    col="time",
    style = "smoker"
)
```

```python

fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.relplot(
    data = fmri,
    x = 'timepoint',
    y = 'signal',
    kind = 'line',
    height = 5,
    col = 'region',
)
#confidence interval
```

```python
sns.relplot(
    data = fmri,
    x = 'timepoint',
    y = 'signal',
    kind = 'line',
    height = 5,
    col = 'region',
    row = 'event'
)
```

```python
sns.relplot(
    data = fmri,
    x = 'timepoint',
    y = 'signal',
    kind = 'line',
    height = 5,
    col = 'region',
    row = 'event',
    estimator = None
)
```
