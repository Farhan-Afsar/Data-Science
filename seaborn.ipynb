```python
penguin = sns.load_dataset('penguins')
penguin.head()
```

```python
sns.displot(
    data = penguin,
    x = 'flipper_length_mm',
    bins = 20,
)
```

```python
sns.displot(
    data = penguin,
    x = 'species',
    discrete = True,
    shrink = 0.7,

)
```

```python
sns.displot(
    data = penguin,
    x = 'flipper_length_mm',
    hue = 'species',
    col = 'species',
)
```

```python
sns.displot(
    data = penguin,
    x = 'flipper_length_mm',
    hue = 'species',
    element = 'step'
)
```

```python
sns.displot(
    data = penguin,
    x = 'flipper_length_mm',
    hue = 'species',
    multiple = 'stack'
)
```

```python
sns.displot(
    data = penguin,
    x = 'flipper_length_mm',
    hue = 'species',
    col = 'species',
    stat = 'probability',
)
```

```python
sns.displot(
    data = penguin,
    x = 'flipper_length_mm',
    kind = 'kde',
    fill = True,
    hue = 'species',
)
```

```python
sns.displot(
    data = penguin,
    x = 'flipper_length_mm',
    kde = True,
    bins = 10,
    shrink = 0.9,
)
```

```python
sns.displot(
    data = penguin,
    x ='bill_length_mm',
    y = 'bill_depth_mm',
    cbar = True
)
```

```python
sns.pairplot(penguin)
```

```python
g = sns.PairGrid(penguin)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot,fill=True)
g.map_diag(sns.histplot,kde=True)
```

```python
'''Categorical Plot'''

tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['day'].value_counts()
```

```python
sns.catplot(
    data=tips,
    x = 'day',
    y = 'total_bill',
    kind = 'strip',
)
```

```python
sns.catplot(
    data=tips,
    x = 'day',
    y = 'total_bill',
    kind = 'strip',
    jitter = False,
    order = ['Sun','Sat','Thur','Fri']
)
```

```python
sns.catplot(
    data=tips,
    x = 'day',
    y = 'total_bill',
    kind = 'swarm',
    order = ['Sun','Sat','Thur','Fri']
)
```

```python
sns.catplot(
    data=tips,
    x = 'day',
    y = 'total_bill',
    kind = 'swarm',
    order = ['Sun','Sat','Thur','Fri'],
    hue = 'smoker',
    col = 'smoker'
)
```

```python
sns.catplot(
    data=tips,
    x = 'day',
    y = 'total_bill',
    kind = 'box',

)
```

```python
sns.catplot(
    data=tips,
    x = 'day',
    y = 'total_bill',
    kind = 'box',
    col = 'smoker',
    hue = 'smoker',
)
```

```python
sns.catplot(
    data=tips,
    x = 'day',
    y = 'total_bill',
    kind = 'boxen',
)
```

```python
# Boxplot + KDE

sns.catplot(
    data=tips,
    x = 'day',
    y = 'total_bill',
    kind = 'violin',
    hue = 'smoker',
    col = 'smoker',

)
```

```python
sns.catplot(
    data=tips,
    x = 'day',
    kind = 'count',
)
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.catplot(
    data=titanic,
    x = 'sex',
    y = 'survived',
    kind = 'point',
)
```

```python
sns.catplot(
    data=titanic,
    x = 'sex',
    y = 'age',
    kind = 'point',
)
```

```python
sns.catplot(
    data=titanic,
    x='class',
    y= 'survived',
    kind = 'point',
    hue = 'sex',
    palette = {
        'male': 'green',
        'female': 'red',
    },
    markers = ['^', 'x'],
    linestyles = ['solid', 'dashed'],

)
```
